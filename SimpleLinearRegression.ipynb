```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
data = {'Hours': [1, 2, 3, 4, 5], 'Marks': [35, 50, 65, 70, 85]}
df = pd.DataFrame(data)

# Features and target
X = df[['Hours']]
y = df['Marks']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict
predicted = model.predict(X)

# Plot
plt.scatter(X, y, color='blue')
plt.plot(X, predicted, color='red')
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.title("Linear Regression: Study vs Marks")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# New sample dataset
data = {
    'YouTube Videos Watched': [1, 3, 5, 7, 9, 11],
    'Understanding Score': [45, 50, 60, 65, 75, 80]
}
df = pd.DataFrame(data)

# Features and target
X = df[['YouTube Videos Watched']]
y = df['Understanding Score']

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict
predicted = model.predict(X)

# Plot
plt.scatter(X, y, color='green')
plt.plot(X, predicted, color='orange')
plt.xlabel("YouTube Videos Watched")
plt.ylabel("Understanding Score")
plt.title("Linear Regression: Videos Watched vs Understanding")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler

# Complex dataset
data = {
    'Study Hours': [2, 4, 6, 8, 10, 12, 14, 16],
    'Practice Problems': [10, 20, 40, 60, 80, 100, 120, 140],
    'Mentorship Sessions': [1, 1, 2, 2, 3, 3, 4, 5],
    'Final Score': [50, 55, 60, 70, 75, 80, 90, 95]
}
df = pd.DataFrame(data)

# Features and target
X = df[['Study Hours', 'Practice Problems', 'Mentorship Sessions']]
y = df['Final Score']

# Optional: Standardize features for better scale (optional here)
# scaler = StandardScaler()
# X = scaler.fit_transform(X)

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict
predicted = model.predict(X)

# Visualization: Study Hours vs Final Score (2D)
plt.figure(figsize=(8, 5))
plt.scatter(df['Study Hours'], y, color='blue', label='Actual')
plt.scatter(df['Study Hours'], predicted, color='red', label='Predicted')
plt.xlabel("Study Hours")
plt.ylabel("Final Exam Score")
plt.title("Regression: Study Hours vs Final Score")
plt.legend()
plt.grid(True)
plt.show()
```
